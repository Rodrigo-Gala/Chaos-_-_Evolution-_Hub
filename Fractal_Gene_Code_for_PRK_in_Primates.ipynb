## **CONTEO FRECUENCIA RELATIVA  Y TOTAL DE DNA**

---



```python
def calculate_base_content(dna_sequence):
    length = len(dna_sequence)
    a_count = dna_sequence.count('A')
    t_count = dna_sequence.count('T')
    g_count = dna_sequence.count('G')
    c_count = dna_sequence.count('C')
    total_bases = a_count + t_count + g_count + c_count
    a_percent = (a_count / total_bases) * 100
    t_percent = (t_count / total_bases) * 100
    g_percent = (g_count / total_bases) * 100
    c_percent = (c_count / total_bases) * 100
    return a_percent, t_percent, g_percent, c_percent

def plot_base_content(a_percent, t_percent, g_percent, c_percent, species_name,save_path):
    bases = ['A', 'T', 'G', 'C']
    percentages = [a_percent, t_percent, g_percent, c_percent]
    colors = ['green', 'red', 'black', 'blue']

    plt.figure(figsize=(8, 6))
    bars = plt.bar(bases, percentages, color=colors)
    plt.title(f'Relative Frequency of bases in {species_name} gene')
    plt.xlabel('Bases')
    plt.ylabel('Relative Percentage (%)')
    plt.ylim(0, 100)

    for bar, percentage in zip(bars, percentages):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{percentage:.2f}%',
                 ha='center', va='bottom')

    plt.savefig(save_path)
    plt.show()
```

## **TOTALES**

```python
def calculate_base_content_T(dna_sequence):
    length = len(dna_sequence)
    a_count = dna_sequence.count('A')
    t_count = dna_sequence.count('T')
    g_count = dna_sequence.count('G')
    c_count = dna_sequence.count('C')
    total_bases = length
    return a_count, t_count, g_count, c_count

def plot_base_content_T(a_count, t_count, g_count, c_count, species_name, save_path):
    bases = ['A', 'T', 'G', 'C']
    counts = [a_count, t_count, g_count, c_count]
    colors = ['green', 'red', 'black', 'blue']

    plt.figure(figsize=(8, 6))
    bars = plt.bar(bases, counts, color=colors)
    plt.title(f'Total Frequency of bases in {species_name} gene')
    plt.xlabel('Bases')
    plt.ylabel('Total Count')

    for bar, count in zip(bars, counts):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{count}',
                 ha='center', va='bottom')

    # Guardar la imagen en la ruta especificada
    plt.savefig(save_path)
    plt.show()

```

## **CONTEO DE K MER Y GENERACION DE MATRIZ**

```python
# Define la función count_kmers
def count_kmers(seq, k):
    d = collections.defaultdict(int)
    for i in range(len(seq) - (k - 1)):
        d[seq[i:i+k]] += 1
    return d

# Define la función chaos_game_representation
def chaos_game_representation(frequencies, k):
    array_size = int(math.sqrt(4**k))
    chaos = []
    for i in range(array_size):
        chaos.append([0] * array_size)

    maxx = array_size
    maxy = array_size
    posx = 1
    posy = 1
    for key, value in frequencies.items():
        for char in key:
            if char == "A":
                posx += maxx / 2
            elif char == "T":
                posy += maxy / 2
            elif char == "G":
                posx += maxx / 2
                posy += maxy / 2
            maxx = maxx / 2
            maxy /= 2
        chaos[int(posy)-1][int(posx)-1] = value
        maxx = array_size
        maxy = array_size
        posx = 1
        posy = 1

    return chaos
```

## **COMPARATIVA DE MATRICES**

```python
def compare_matrices(matrices):
    matrix1 = np.array(matrices[0])
    percentage_differences = []

    for matrix in matrices[1:]:
        # Convierte la matriz a un array de NumPy
        matrix = np.array(matrix)

        # Calcula la diferencia absoluta entre las dos matrices
        absolute_difference = np.abs(matrix1 - matrix)

        # Calcula el porcentaje de diferencia en comparación con la primera matriz
        percentage_difference = (np.sum(absolute_difference) / np.sum(matrix1)) * 100
        percentage_differences.append(percentage_difference)

    return percentage_differences
```

```python
import os
import matplotlib.pyplot as plt

# Lista de tus conjuntos de datos
conjuntos_de_datos = [d1_7, d2_7, d3_7, d4_7, d5_7, d6_7, d7_7, d8_7,
                      d9_7, d10_7, d11_7, d12_7, d13_7, d14_7, d15_7, d16_7]

nombres_especies = [
    r'$\mathit{H. sapiens}$',
    r'$\mathit{G. gorilla \ PRKN}$',
    r'$\mathit{P. paniscus \ PRKN}$',
    r'$\mathit{P. troglodytes \ PRKN}$',
    r'$\mathit{P. abelii \ PRKN}$',
    r'$\mathit{P. pygmaeus \ PRKN}$',
    r'$\mathit{M. fascicularis \ PRKN}$',
     r'$\mathit{M. mulatta \ PRKN}$',
    r'$\mathit{M. thibetana \ PRKN}$',
    r'$\mathit{R. roxellana \ PRKN}$',
    r'$\mathit{T. francoisi \ PRKN}$',
    r'$\mathit{N. leucogenys \ PRKN}$',
    r'$\mathit{S. syndactylus \ PRKN}$',
    r'$\mathit{M. murinus}$',
    r'$\mathit{C. jacchus}$',
    r'$\mathit{L. catta}$',
]

# Lista de nombres de especies correspondientes a cada conjunto de datos
etiquetas_especies = [
    '01 H. sapiens',
    '02 G. gorilla',
    '03 P. paniscus',
    '04 P. troglodytes',
    '05 P. abelii',
    '06 P. pygmaeus',
    '07 M. fascicularis',
    '08 M. mulatta',
    '09 M. thibetana',
    '10 R. roxellana',
    '11 T. francoisi',
    '12 N. leucogenys',
    '13 S. syndactylus',
    '14 M. murinus',
    '15 C. jacchus',
    '16 L. catta',
]

# Ruta base de la carpeta donde se guardarán las imágenes
ruta_base = "/content/drive/MyDrive/TESIS_FINAL/Suplementary/02 Supplementari material/K-mer nucleotide appearance frequency and composition/"

# Para cada conjunto de datos, genera y guarda el gráfico
for i, datos in enumerate(conjuntos_de_datos, start=1):
    # Ruta de la carpeta específica para esta especie
    carpeta_especie = os.path.join(ruta_base, etiquetas_especies[i-1])

    # Si la carpeta no existe, se crea
    if not os.path.exists(carpeta_especie):
        os.makedirs(carpeta_especie)

    # Cambiar el directorio actual a la carpeta de la especie
    os.chdir(carpeta_especie)

    # Crear una nueva figura con el tamaño deseado
    plt.figure(figsize=(20, 11))
    # Calcular frecuencias y ordenar por frecuencia
    freq_list = sorted(list(datos.values()), reverse=True)

    # Crear un gráfico de barras
    plt.bar(
        range(1, len(freq_list) + 1),
        freq_list,
        label=f'{nombres_especies[i-1]}',  # Usar el nombre de la especie en lugar de "Datos"
        color='blue',    # Color de las barras
        alpha=0.7        # Transparencia
    )

    # Configurar etiquetas y título
    plt.xlabel('Range')
    plt.ylabel('Frequency')
    plt.title(f'Frequency Plot for {nombres_especies[i-1]} of size 7-mers')

    # Mostrar leyenda
    plt.legend()

    # Guardar el gráfico con el nombre deseado
    nombre_archivo = etiquetas_especies[i-1] + "_k7.png"
    plt.savefig(nombre_archivo)

    # Cerrar la figura para liberar memoria
    plt.close()

    # Cambiar el directorio de trabajo de vuelta al original
    os.chdir(ruta_base)
```

```python
def entropy(sequence):
    # Calcula la distribución de probabilidad de los caracteres en la secuencia.
    counts = collections.Counter(sequence)
    probabilities = {char: count / len(sequence) for char, count in counts.items()}

    # Calcula la entropía de Shannon.
    entropy = 0
    for probability in probabilities.values():
        if probability > 0:
            entropy -= probability * math.log2(probability)

    return round(entropy, 5)  # Reducir el resultado a 4 decimales
```

```python
def generate_points(dna_sequence):
    # Define las esquinas del cuadrado
    corners = {'G': (0, 100), 'A': (100, 100), 'T': (0, 0), 'C': (100, 0)}

    # Punto inicial
    current_point = (0.5, 0.5)

    # Guarda las coordenadas de los puntos para graficar
    x_values = [current_point[0]]
    y_values = [current_point[1]]

    # Itera sobre la secuencia de ADN
    for nucleotide in dna_sequence:
        # Obtiene la coordenada de la esquina correspondiente al nucleótido actual
        target_corner = corners[nucleotide]

        # Calcula el punto a la mitad de la distancia entre el punto actual y la esquina objetivo
        current_point = ((current_point[0] + target_corner[0]) / 2, (current_point[1] + target_corner[1]) / 2)

        # Guarda las coordenadas del nuevo punto
        x_values.append(current_point[0])
        y_values.append(current_point[1])

    return x_values, y_values
```

```python
def normalize_scale(data, reference_scale):
    min_data = min(data)
    max_data = max(data)
    normalized_data = [(x - min_data) / (max_data - min_data) * (reference_scale[1] - reference_scale[0]) + reference_scale[0] for x in data]
    return normalized_data
```
