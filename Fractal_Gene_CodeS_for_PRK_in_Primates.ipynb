## **FCGR NUMERICAL MATRIX GENERATOR**

```python
# Define la función chaos_game_representation
def chaos_game_representation(frequencies, k):
    array_size = int(math.sqrt(4**k))
    chaos = []
    for i in range(array_size):
        chaos.append([0] * array_size)

    maxx = array_size
    maxy = array_size
    posx = 1
    posy = 1
    for key, value in frequencies.items():
        for char in key:
            if char == "A":
                posx += maxx / 2
            elif char == "T":
                posy += maxy / 2
            elif char == "G":
                posx += maxx / 2
                posy += maxy / 2
            maxx = maxx / 2
            maxy /= 2
        chaos[int(posy)-1][int(posx)-1] = value
        maxx = array_size
        maxy = array_size
        posx = 1
        posy = 1

    return chaos
```

```python
import collections
import math
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def entropy(sequence):
    # Calculate the probability distribution of characters in the sequence.
    counts = collections.Counter(sequence)
    probabilities = {char: count / len(sequence) for char, count in counts.items()}

    # Calculate Shannon entropy.
    entropy = 0
    for probability in probabilities.values():
        if probability > 0:
            entropy -= probability * math.log2(probability)

    return round(entropy, 5)  # Reduce the result to 5 decimal places


# Example entropy calculations (replace DATOS1, DATOS2, etc. with actual data sequences)
entropy1 = entropy(DNA_SEQUENCE1)
entropy2 = entropy(DNA_SEQUENCE2)
entropy3 = entropy(DNA_SEQUENCE3)
entropy4 = entropy(DNA_SEQUENCE4)
entropy5 = entropy(DNA_SEQUENCE5)
entropy6 = entropy(DNA_SEQUENCE6)
entropy7 = entropy(DNA_SEQUENCE7)
entropy8 = entropy(DNA_SEQUENCE8)
entropy9 = entropy(DNA_SEQUENCE9)
entropy10 = entropy(DNA_SEQUENCE10)
entropy11 = entropy(DNA_SEQUENCE11)
entropy12 = entropy(DNA_SEQUENCE12)
entropy13 = entropy(DNA_SEQUENCE13)
entropy14 = entropy(DNA_SEQUENCE14)
entropy15 = entropy(DNA_SEQUENCE15)
entropy16 = entropy(DNA_SEQUENCE16)

# Desired path
desired_path = "/path/to/your/directory"

# Entropy data
entropies = {
    'H. sapiens': entropy1,
    'G. gorilla': entropy2,
    'P. paniscus': entropy3,
    'P. troglodytes': entropy4,
    'P. abeli': entropy5,
    'P. pygmaeus': entropy6,
    'M. fascicularis': entropy7,
    'M. mulatta': entropy8,
    'M. thibetana': entropy9,
    'R. roxellana': entropy10,
    'T. francoisi': entropy11,
    'N. leucogenys': entropy12,
    'S. syndactylus': entropy13,
    'M. murinus': entropy14,
    'C. jacchus': entropy15,
    'L. catta': entropy16,
}

# Create a pandas DataFrame
df = pd.DataFrame(list(entropies.items()), columns=['Primate species', 'Shannon entropy'])

# Sort the DataFrame by entropy from lowest to highest
df = df.sort_values(by='Shannon entropy')

# Full path to save the Excel file
excel_path = os.path.join(desired_path, 'Shannon_entropy_Primate_species.xlsx')

# Save the DataFrame to an Excel file with bold and italic species names
with pd.ExcelWriter(excel_path, engine='xlsxwriter') as writer:
    df.style.applymap(lambda x: 'font-weight: bold; font-style: italic', subset=['Primate species']).to_excel(writer, index=False, sheet_name='Entropies')

# Show success message
print(f"Excel file saved at: {excel_path}")

# Define the destination path
destination_path = "/path/to/your/directory"

# Sort the entropies from lowest to highest
sorted_entropies = sorted(entropies.items(), key=lambda x: x[1])

# Extract the species names and sorted entropies
sorted_species, sorted_entropies = zip(*sorted_entropies)

# Calculate the mean and standard deviation
mean_entropy = np.mean(sorted_entropies)
std_deviation = np.std(sorted_entropies)

# Plot the entropies
plt.figure(figsize=(13, 9))
plt.plot(sorted_species, sorted_entropies, marker='o', color='blue', linestyle='-')

# Add line for the mean
plt.axhline(y=mean_entropy, color='red', linestyle='--', label=f'Average: {mean_entropy:.5f}')

# Add lines for the standard deviation
plt.axhline(y=mean_entropy + std_deviation, color='green', linestyle='--', label=f'Average + Standard deviation: {mean_entropy + std_deviation:.5f}')
plt.axhline(y=mean_entropy - std_deviation, color='green', linestyle='--', label=f'Average - Standard deviation: {mean_entropy - std_deviation:.5f}')

# Label the points with the entropy values
for i, txt in enumerate(sorted_entropies):
    plt.annotate(f'{txt:.5f}', (i, sorted_entropies[i]), textcoords="offset points", xytext=(-4,-4), ha='left', va='top', rotation=340, fontsize=8)

# Set the x-axis with the species names
plt.xticks(range(len(sorted_species)), sorted_species, rotation=45, ha='right')
plt.ylabel('Shannon Entropy', fontweight='bold')
plt.title(r'Shannon Entropy for Different $\mathit{PRKN}$ Gene from Primate Species')

# Add a legend
plt.legend()

# Save the plot to the destination path
if not os.path.exists(destination_path):
    os.makedirs(destination_path)
plt.savefig(os.path.join(destination_path, 'Entropy_primate_species.png'))

# Show the plot
plt.tight_layout()
plt.show()
```

# **Software Caos Game Representation (*CGR*) - CODE DNA Chaos Game Algorithm (*DCGA*)**


```python
import os
import matplotlib.pyplot as plt
import gc

def generate_points(dna_sequence):
    # Define the corners of the square
    corners = {'G': (0, 100), 'A': (100, 100), 'T': (0, 0), 'C': (100, 0)}

    # Initial point
    current_point = (0.5, 0.5)

    # Store the coordinates of the points for plotting
    x_values = [current_point[0]]
    y_values = [current_point[1]]

    # Iterate over the DNA sequence
    for nucleotide in dna_sequence:
        # Get the corner coordinate corresponding to the current nucleotide
        target_corner = corners[nucleotide]

        # Calculate the point halfway between the current point and the target corner
        current_point = ((current_point[0] + target_corner[0]) / 2, (current_point[1] + target_corner[1]) / 2)

        # Store the coordinates of the new point
        x_values.append(current_point[0])
        y_values.append(current_point[1])

    return x_values, y_values

# Example data
dna_data = "your_dna_sequence_here"
output_path = "your_output_path_here"
figure_title = "your_title_here"
figure_dpi = 300
figure_size = (8, 6)
point_size = 1
marker_style = 'o'
color_map = 'viridis'

x1, y1 = generate_points(dna_data)

# Normalize point indices using the reference scale
reference_scale = (0, 100)
normalized_indices = range(len(x1))

# Define corners with labels
corners_labels = {'G': (0, 100), 'A': (100, 100), 'T': (0, 0), 'C': (100, 0)}

# Plot only the points without lines, with reduced size
fig, ax = plt.subplots(figsize=figure_size, dpi=figure_dpi)
scatter_plot = plt.scatter(x1, y1, c=normalized_indices, cmap=color_map, s=point_size, marker=marker_style)

# Show corners with labels
for nucleotide, (x, y) in corners_labels.items():
    plt.text(x, y, nucleotide, fontsize=5, ha='center', va='center', color='black')

plt.title(f"CGR of {figure_title}")
plt.xlabel("")
plt.ylabel("")

# Add the color bar
colorbar = plt.colorbar(scatter_plot, label='Point index')

# Save the figure
plt.savefig(os.path.join(output_path, f"CGR_of_{figure_title}.png"), dpi=figure_dpi, bbox_inches='tight')

plt.show()
gc.collect()
```

```python
7
import os
import cv2
import numpy as np

# Folder containing the images
folder = "/path/to/images"

# Check if the folder exists
if os.path.exists(folder):
    # Get the list of files in the folder
    files = os.listdir(folder)

    # Initialize a list to store the image paths
    paths = []

    # Iterate over each file and get the full path
    for file in files:
        full_path = os.path.join(folder, file)
        # Check if the file is an image based on its extension
        if any(file.lower().endswith(ext) for ext in ['.jpg', '.jpeg', '.png', '.gif', '.bmp']):
            paths.append(full_path)
else:
    print(f"The folder {folder} does not exist.")

# Initialize a dictionary to store the comparison results
comparison_results = {}

# Iterate over the reference images
for i, reference in enumerate(paths):
    # Load the reference image
    reference_image = cv2.imread(reference)

    # Ensure that the reference image is not None
    if reference_image is None:
        print(f"Could not load the reference image at {reference}")
        continue

    # Calculate the total number of pixels in the reference image
    total_reference_pixels = reference_image.shape[1] * reference_image.shape[0]

    # Iterate over the comparison images
    for j, comparison in enumerate(paths):
        # Load the comparison image
        comparison_image = cv2.imread(comparison)

        # Ensure that both images have the same dimensions
        comparison_image = cv2.resize(comparison_image, (reference_image.shape[1], reference_image.shape[0]))

        # Compare the images pixel by pixel and get the difference
        difference = cv2.absdiff(reference_image, comparison_image)

        # Get the binary color intensity (black or white)
        _, binary = cv2.threshold(difference, 30, 255, cv2.THRESH_BINARY)

        # Calculate the number of different pixels
        different_pixels = np.sum(binary == 255)

        # Calculate the binary color intensity (percentage of different pixels)
        binary_intensity = (different_pixels / total_reference_pixels) * 100

        # Store the comparison result in the comparison results dictionary
        key = f"ID_{i+1}vs_{j+1}"
        comparison_results[key] = binary_intensity

# Dynamically create variables for each combination of ID and store the corresponding value
for key, value in comparison_results.items():
    globals()[key] = value
```

# **CODE Complementary Nucleotide Sequence Mutation (*CNSMut*)**

```python
import random
import math

def generate_mutated_sequence(original_sequence, num_changes):
    mutated_sequence = list(original_sequence)
    bases = ['A', 'T', 'C', 'G']

    # Get unique positions to make changes
    change_positions = random.sample(range(len(original_sequence)), num_changes)

    # Change the bases at the selected positions
    for pos in change_positions:
        original_base = original_sequence[pos]
        bases_without_orig = [b for b in bases if b != original_base]
        mutated_sequence[pos] = random.choice(bases_without_orig)

    return ''.join(mutated_sequence)

# Number of changes to make in each sequence
num_changes = n

# List to store the generated sequences
generated_sequences = []

for _ in range(num_mutated_sequences):
    generated_sequences.append(generate_mutated_sequence(original_sequence, num_changes))

# Individual assignment to variables
for i, sequence in enumerate(generated_sequences, start=1):
    globals()[f"SEQUENCE_{i}"] = sequence

# Generate mutated sequences and calculate entropy for each one
entropies = []
for _ in range(num_mutated_sequences):
    mutated_sequence = generate_mutated_sequence(original_sequence, num_changes)
    generated_sequences.append(mutated_sequence)
    entropies.append(entropy(mutated_sequence))

# Calculate the average entropy
average_entropy = sum(entropies) / len(entropies)

# Calculate the standard deviation of the average entropy
std_dev_entropy = math.sqrt(sum((x - average_entropy) ** 2 for x in entropies) / len(entropies))

print("Average entropy:", average_entropy)
print("Standard deviation of the average entropy:", std_dev_entropy)
```

# **CODE Chemical Swap Entropy Analysis (*ChemSwap*)**

```python
import random
import math

def generate_mutated_sequence(original_sequence, num_changes):
    mutated_sequence = list(original_sequence)
    purine_bases = ['A', 'G']  # Purines
    pyrimidine_bases = ['T', 'C']  # Pyrimidines

    for _ in range(num_changes):
        index = random.randint(0, len(mutated_sequence)-1)
        original_base = mutated_sequence[index]
        if original_base in purine_bases:
            new_base = random.choice(pyrimidine_bases)
        elif original_base in pyrimidine_bases:
            new_base = random.choice(purine_bases)
        else:
            # In case it is neither a purine nor a pyrimidine, no change is made
            continue

        mutated_sequence[index] = new_base

    return ''.join(mutated_sequence)

# Number of changes to make in each sequence
num_changes = n

# List to store generated sequences
generated_sequences = []

for _ in range(num_mutated_sequences):
    generated_sequences.append(generate_mutated_sequence(original_sequence, num_changes))

# Individual assignment to variables
for i, sequence in enumerate(generated_sequences, start=1):
    globals()[f"DATA1_G1_{i}"] = sequence

# Generate mutated sequences and calculate entropy for each
entropies = []
for _ in range(num_mutated_sequences):
    mutated_sequence = generate_mutated_sequence(original_sequence, num_changes)
    generated_sequences.append(mutated_sequence)
    entropies.append(entropy(mutated_sequence))

# Calculate the average entropy
average_entropy = sum(entropies) / len(entropies)

# Calculate the standard deviation of the average entropy
entropy_std_dev = math.sqrt(sum((x - average_entropy) ** 2 for x in entropies) / len(entropies))

print("Average entropy:", average_entropy)
print("Standard deviation of the average entropy:", entropy_std_dev)
```
